```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

X = pd.DataFrame(df['Gender'])
y = pd.DataFrame(df['Telepresence'])

model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
 model.fit(X.iloc[train,:], y.iloc[train,:])
 score = model.score(X.iloc[test,:], y.iloc[test,:])
 scores.append(score)
print(scores)
```

```python
# Here we see that R^2 is 0.36 which is a quite good result for simple linear regression, therefore we can say that Gender has a negative influence on importance of Telepresence which means that Female audience is more likely too pay attention to this stimulus. 
```
